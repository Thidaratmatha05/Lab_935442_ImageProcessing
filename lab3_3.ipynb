```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('/content/1.jpg', 0)
sobel = cv2.Sobel(img,cv2.CV_64F,1,1,ksize=5) #min=3 -- max= 5
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)  # x10 y11

plt.subplot(221),plt.imshow(sobelx, cmap="gray")
plt.subplot(222),plt.imshow(sobely, cmap="gray")
plt.subplot(223),plt.imshow(sobel, cmap="gray")
plt.subplot(224),plt.imshow(img, cmap="gray")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('/content/1.jpg', 0)
f = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
f1 = np.array([[1,0,-1], [1,0,-1], [1,0,-1]])
f2 = np.array([[-1,-2,-1], [0,0,0], [1,2,1]])

sharp = cv2.filter2D(img, -1, f)
sharp1 = cv2.filter2D(img, -1, f1)
sharp2 = cv2.filter2D(img, -1, f2)

plt.subplot(221),plt.imshow(sharp, cmap="gray")
plt.subplot(222),plt.imshow(sharp1, cmap="gray")
plt.subplot(223),plt.imshow(sharp2, cmap="gray")
plt.subplot(224),plt.imshow(img, cmap="gray")
plt.show()
```
